```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Practice-Datasets .csv")
df
```

```python
#histogram
plt.figure(figsize=(10,6))
plt.hist(df['Age'], bins=10)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age Distribution")
plt.show()
```

```python
#piechart
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Percentage Distribution of Gender")
plt.show()
```

```python
plt.figure(figsize=(7,8))
df["City"].value_counts().plot(kind="bar")
plt.title("Customers per City")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['Monthly_Income'])
plt.ylabel("Monthly Income")
plt.title("Box Plot of Monthly Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.hist(df['Spending_Score'],color='pink', bins=10)
plt.xlabel("Spending Score")
plt.ylabel("Frequency")
plt.title("Histogram of Spending Score Distribution")
plt.show()
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df = df.sort_values('Purchase_Date')
# Plot Line Chart
plt.figure()
plt.plot(df['Purchase_Date'], df['Purchase_Amount'],color='orange')
plt.xlabel("Purchase Date")
plt.ylabel("Purchase Amount")
plt.title("Purchase Amount Trend Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_income = df.groupby('Gender')['Monthly_Income'].mean()
# Create Bar Plot
plt.figure()
plt.bar(avg_income.index, avg_income.values,color='red')
plt.xlabel("Gender")
plt.ylabel("Average Monthly Income")
plt.title("Average Monthly Income by Gender")
plt.show()
```

```python
# Create box plot grouped by Product Category
df.boxplot(column='Purchase_Amount', by='Product_Category')
plt.title("Purchase Amount by Product Category")
plt.suptitle("")  # removes default extra title
plt.xlabel("Product Category")
plt.ylabel("Purchase Amount")
plt.show()
```

```python
# Product Category percentage
plt.figure()
df["Product_Category"].value_counts().plot.pie(
    autopct="%1.1f%%",
    colors=["pink","red","lightgreen","salmon","violet"]
)
plt.ylabel("")
plt.show()
```

```python
# Histogram of Purchase Amount
plt.figure()
plt.hist(df["Purchase_Amount"], bins=20, color="brown")
plt.show()
```

```python
# Average Spending Score by City
plt.figure()
df.groupby("City")["Spending_Score"].mean().plot.bar(color="teal")
plt.show()
```

```python
# Monthly total purchase amount
monthly = df.resample('M', on='Purchase_Date')["Purchase_Amount"].sum()
plt.figure()
plt.plot(monthly, color="darkgreen")
plt.show()
```

```python
# Monthly Income comparison across cities
plt.figure()
df.boxplot(column="Monthly_Income", by="City",
           patch_artist=True,
           boxprops=dict(facecolor="lightyellow"))
plt.suptitle("")
plt.show()
```

```python
# Age distribution by gender
plt.figure()
for g,color in zip(df["Gender"].unique(), ["blue","red"]):
    plt.hist(df[df["Gender"]==g]["Age"], bins=15, alpha=0.5, label=g, color=color)

plt.legend()
plt.show()
```

```python
#Total purchase by product category
plt.figure()
df.groupby("Product_Category")["Purchase_Amount"].sum().plot.bar(color="magenta")
plt.show()
```

```python
# Relationship between Monthly Income and Spending Score
plt.figure()
plt.scatter(df["Monthly_Income"], df["Spending_Score"], color="darkorange")
plt.show()
```

```python
# Daily purchase amount variation
daily = df.groupby(df["Purchase_Date"].dt.date)["Purchase_Amount"].sum()
plt.figure()
plt.plot(daily, color="navy")
plt.show()
```

```python
# Spending Score distribution by gender
plt.figure()
df.boxplot(column="Spending_Score", by="Gender",
           patch_artist=True,
           boxprops=dict(facecolor="lightpink"))
plt.suptitle("")
plt.show()
```

```python
# Customer count by product category and gender
plt.figure()
pd.crosstab(df["Product_Category"], df["Gender"]).plot.bar(color=["green","purple"])
plt.show()
```

```python
# Relationship between Age and Purchase Amount
plt.figure()
plt.scatter(df["Age"], df["Purchase_Amount"], color="pink")
plt.show()
```
